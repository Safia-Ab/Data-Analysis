```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query('select * from sqlite_master', conn)
```

```python
pd.read_sql_query('select * from salaries', conn)
```

```python
def sql_query(q):
    return pd.read_sql_query(q,conn)
```

#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

```python
q = '''select jobtitle, Avg(TotalPayBenefits) Avg_TPayB
    from Salaries
    Where jobtitle Not In ('Not Provided', 'Not provided') Group By jobtitle
    Order by 1 '''

sql_query(q)
```

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

```python
q = ''' select Jobtitle, OvertimePay from Salaries
        Where OvertimePay > 500000
        Order by 2 '''
sql_query(q)
```

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

```python
q = ''' select Count(Distinct EmployeeName) as UniqueEN from salaries
        where EmployeeName NOT IN ('Not Provided', 'Not provided') '''
sql_query(q)
```

#### **4.Show The Top 10 highest Paid Employees**

```python
q = ''' select Employeename, Jobtitle, TotalpayBenefits from Salaries
        where Employeename Not In ('Not Provided', 'Not provided')
        Order by 2 Desc Limit 10 '''
sql_query(q)
```

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
q = ''' select EmployeeName,JobTitle,TotalPayBenefits
        from salaries where EmployeeName NOT IN ('Not Provided', 'Not provided')
        Order by 3 Desc '''
sql_query(q)
```
