```python
import pandas as pd

df = pd.read_csv("retail_supermarket.csv")
df.head()
```

```python
print("\n1 Sales > 500 & Quantity > 5:\n", df.query("Sales > 500 and Quantity > 5"))
```

```python
print("\n Discount > 0.5 or Profit < 0:\n", df.query("Discount > 0.5 or Profit < 0"))
```

```python
print("\n East region (Ship Mode & Profit):\n", df.loc[df.Region == "East", ["Ship Mode", "Profit"]])
```

```python
print("\n Corporate + Central count:", len(df.query("Segment == 'Corporate' and Region == 'Central'")))
```

```python
print("\n Furniture but not Chairs:\n", df.query("Category == 'Furniture' and `Sub-Category` != 'Chairs'"))
```

```python
print("\n  Discount > 0.3 and Profit > 0:\n", df.query("Discount > 0.3 and Profit > 0"))
```

```python
print("\n Technology, Quantity > 5, Profit > 0:\n", df.query("Category == 'Technology' and Quantity > 5 and Profit > 0"))
```

```python
print("\n Rows 04, selected columns:\n", df.iloc[:5, [df.columns.get_loc('Ship Mode'), df.columns.get_loc('Segment'), df.columns.get_loc('Sales')]])
```

```python
print("\n Home Office, Sales > 1000, Quantity > 3:\n", df.query("Segment == 'Home Office' and Sales > 1000 and Quantity > 3"))
df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
print("\n Profit Margin > 0.3:\n", df.query("`Profit Margin` > 0.3"))
```
