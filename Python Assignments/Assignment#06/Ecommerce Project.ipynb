```python
df.fillna({'City' : 3.0, "order Priority" :2.0})
```

```python
#sales = Price * Quantity
sales = df['price'] * df['quantity']
sales
```

```python
df['sales'] = (df['price'] * df['quantity']) / 1000
df['sales']
```

```python
df1 = df.groupby('product')['sales'].sum().sort_values(ascending=False).head(5).reset_index()
df1
```

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='product', y='sales', data=df1, palette='viridis')
for i, v in enumerate(df1['sales']):
    plt.text(i, v + 0.05, f'{v:.2f}K', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Best Selling Products', fontsize=12, color= 'blue')
plt.xlabel('Total Sales (in thousands)', fontsize=12)
plt.ylabel('Product', fontsize=12)
plt.show()
print("Observation:")
print("Top selling products are Headphones, Watch and Mobile")
```

```python
df2 = df['payment_method'].value_counts()
df2
```

```python
Values2 = df2.to_list()
Values2
```

```python
plt.figure(figsize=(6, 6))
plt.pie(x = Values2, labels= df2.index, autopct='%1.1f%%', startangle=90)
plt.title('Payment Method Distribution', fontsize=12, color= 'blue')
plt.axis('equal') 
plt.show()
print("Observation:")
print("Most used payment method is JazzCash", )
```

```python
month_sales = df.groupby('Month')['total'].sum()
month_sales
```

```python
plt.figure(figsize=(7,4))
plt.plot(month_sales.index, month_sales.values, marker='o')
plt.title("Monthwise Sales Trend", color= 'blue')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45) 
plt.show()
print("Observation:")
print("Sales fluctuate across months with peaks around months having highest values.")
```

```python
city_sales = df.groupby('City')['total'].sum().sort_values()
city_sales
```

```python
 city_names = {
    1: 'Karachi',
    2: 'Lahore', 
    3: 'Islamabad',
    4: 'Faisalabad',
    5: 'Quetta',
    6: 'Peshawar'
}
```

```python
plt.figure(figsize=(8,5))
plt.bar(city_sales.index, city_sales.values)
plt.xticks(list(city_names.keys()), list(city_names.values()))
plt.title("Citywise Total Sales", color= 'blue')
plt.xlabel("Total Sales")
plt.ylabel("City")
plt.show()

print("Observation:")
print("Highest sales are recorded in Lahore")
```

```python
cat_month = df.groupby(['Month','category'])['total'].sum().unstack()
cat_month
```

```python
cat_month.plot(kind='bar', figsize=(9,5))
plt.title("Monthly Sales by Category", color= 'blue')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend(title="Category")
plt.xticks(rotation=45) 
plt.show()

print("Observation:")
print("Category sales vary every month.")
```
