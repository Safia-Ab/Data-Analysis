```python
import csv
```

```python
open_file = open('student data1.csv', mode = 'r')
open_file
```

```python
reader = csv.reader(open_file)
reader
```

```python
data = list(reader)
data
```

```python
for g in data[1:]:
    marks = float(g[5])
    if marks >= 80:
        grades = 'A'
    elif marks >= 70:
        grades = 'B'
    elif marks >= 60: 
        grades = 'c'
    elif marks >= 50:
        grades = 'D'
    else: 
        grades = 'F'
    g.append(grades)       
```
